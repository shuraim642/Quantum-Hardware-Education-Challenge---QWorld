Research laboratories is conducting research on a new device that actively utilizes the principles of quantum mechanics, the superconducting qubit. As quantum mechanics is a physical laws that govern the microscopic world of atoms and electrons using superconductivity, this device aims to apply quantum mechanics directly even in micron sized electrical circuits. 

Inside the device is a square circuit comprising an aluminum loop. The three indentations are called Josephson junctions the junctions are comprised of a thin insulator sandwiched by super conductors. 

When this device is cooled to near absolute zero an electric current flows in the loop even if no voltage is applied this phenomenon is known as superconductivity. The current has two states clockwise and counterclockwise normally. 

It would be impossible for clockwise and counterclockwise currents to exist at the same time. But in the world of quantum mechanics however, such a situation is possible this is referred to as a superposition stage.

$$\newcommand{\ket}[1]{\left|{#1}\right\rangle}$$
The magnetic flux that penetrates a super conductive loop is quantized to an integer multiple of the flux quantum. What would happen if we apply two magnetic flux from the outside that is equivalent to half of the flux quantum. We obtain a counterclockwise state ${\ket{L}}$ then intensifies the flux and a clockwise state ${\ket{R}}$ that cancels off the flux.

$$\newcommand{\ket}[1]{\left|{#1}\right\rangle}$$

The **ground state $\ket{0}$** which is energetically the most stable state, is a superposition state of ${\ket{L} + \ket{R}}$. Similarly, the first **excited state $\ket{1}$** is a superposition state of ${\ket{L} - \ket{R}}$.

We need to make and control a lot of qubits in order to build a working quantum computer. Scalability, or the capacity to combine numerous qubits, is constrained since each qubit in a multi-qubit system gradually loses its quantum features. Decoherence is a phenomena that arises from interactions between the qubits and their environment. Utilizing photons is one technique to generate scalable structures (particles of light). We could be onto something since photons' quantum states are more resistant to decoherence. From many approaches, Xanadu focus on linear optical quantum computing, an approach that has already achieved quantum advantage.Since photons seldom interact with one another, it is simple to eliminate uncontrolled interactions that might otherwise change the quantum state of the particle. However, multi-qubit gates are required to create a universal quantum computer, which means that photons must be somehow forced to communicate with one another.

We will initially concentrate on influencing photons using linear materials, whose characteristics do not change when they come into contact with light. A portion of the so-called Gaussian states can be created using linear materials. They are our safest tool in photonics since they can be made with a success record of 100% using regular optical instruments.

```python
import pennylane as qml
from pennylane import numpy as np
```

```python
dev= qml.device("default.qubit", wires=2)
```

```python
@qml.qnode(dev)
def circuit():
    qml.Hadamard(wires=1)
    qml.CNOT(wires=[0,1])
    
    return qml.expval(qml.PauliZ(wires=0))
    return qml.counts()
```

```python
print(qml.draw(circuit, show_matrices=True)())
```

```python
circuit()
```
